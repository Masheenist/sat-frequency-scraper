```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./SatList.csv')
```

```python
myDict = df.to_dict()
```

```python
myFreqs = list(myDict['Frequency [MHz]'].values())
```

```python
newFreq = []
for each in myFreqs:
    try:
        newFreq += [float(each)]
    except:
        pass
```

```python
plt.hist(newFreq, bins=50, range=(0, 3000))
plt.xlabel("Frequency [MHz]")
plt.ylabel("Count")
plt.title("Satellite Downlink Distribution (0 - 3 GHz)")
plt.savefig('SatDist3GHz.png')
#plt.xlim(0, 1000)
#plt.ylim(0, 100)
```

```python
plt.hist(newFreq, bins=50)
plt.xlabel("Frequency [MHz]")
plt.ylabel("Count")
plt.title("Satellite Downlink Distribution (0 - 40 GHz)")
plt.savefig("SatDist40GHz.png")
#plt.xlim(0, 1000)
#plt.ylim(0, 100)
```
